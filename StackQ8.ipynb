```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None

    def is_empty(self):
        return len(self.stack) == 0

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None

    def size(self):
        return len(self.stack)

def is_palindrome(stack):
    if stack.is_empty():
        return True
    
    original_stack = []
    temp_stack = Stack()

    while not stack.is_empty():
        element = stack.pop()
        original_stack.append(element)
        temp_stack.push(element)

    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())

    return original_stack == stack.stack

# Example usage
stack = Stack()
stack.push('a')
stack.push('b')
stack.push('c')
stack.push('b')
stack.push('a')

print("Is the stack a palindrome?", is_palindrome(stack))  # Output: True

stack2 = Stack()
stack2.push('a')
stack2.push('b')
stack2.push('c')

print("Is the stack a palindrome?", is_palindrome(stack2))  # Output: False
```
